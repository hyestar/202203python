```python
import matplotlib.pyplot as plt
# x축 좌표와 y축 좌표를 지정하지 않을 경우 x축은 0부터 1씩 증가 y축은 사용자가 지정한 값으로 인식
plt.style.use('default')
plt.plot([2,1,5,9])
plt.show()
```

```python
# x축 좌표와 y축 좌표를 지정한 경우
x = [2,4,6,8]
y = [1,3,5,7]
plt.plot(x,y)
plt.show()
```
