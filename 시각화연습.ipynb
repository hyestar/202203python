```python
fig = plt.figure()
# fig.add_subplot(행의 갯수, 열의 갯수, 위치값)
# 그래프 위치 지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,3,3)
# 원하는 위치에 그래프를 그림
x = range(0,10)
y = range(5,15)
ax1.plot(x, y, 'bo')
x = range(0, 50)
y = range(0, 50)
ax2.plot(x,y,'y^')
# x축이름, y축이름 출력
ax1.set_xlabel('time')
ax1.set_ylabel('price')
ax2.set_xlabel('time')
ax2.set_ylabel('price')
plt.show()
```

```python
from matplotlib import pyplot as plt
import matplotlib.font_manager as fm
import matplotlib
# 폰트 지정
font_location = "C:\Windows\Fonts\H2PORM.TTF"
fontname = fm.FontProperties(fname=font_location).get_name()
matplotlib.rc('font', family = fontname)

fig = plt.figure()
# fig.add_subplot(행의 갯수, 열의 갯수, 위치값)
# 그래프 위치 지정
ax1 = fig.add_subplot(1,2,1)
ax2 = fig.add_subplot(1,3,3)
# 원하는 위치에 그래프를 그림
x = range(0,10)
y = range(5,15)
ax1.plot(x, y, 'bo')
x = range(0, 50)
y = range(0, 50)
ax2.plot(x,y,'y^')
# x축이름, y축이름 출력
ax1.set_xlabel('시간')
ax1.set_ylabel('가격')
ax2.set_xlabel('시간')
ax2.set_ylabel('가격')
plt.show()
```

```python
# 막대그래프 작성
plt.style.use('ggplot')
fig=plt.figure()
ax = fig.add_subplot(111)
colors = ['red', 'blue']
labels = ['english', 'math']
x = [3,4]
y = [1,2]
ax.bar(x,y, tick_label = labels, color = colors)
for i, v in enumerate(y):
    ax.text(v+2, i+1, str(v), color = 'red')
plt.show()
```

```python
# 수평 막대그래프 작성
plt.style.use('ggplot')
fig=plt.figure()
ax = fig.add_subplot(111)
colors = ['red', 'blue']
labels = ['english', 'math']
x = [3,4]
y = [1,2]
ax.barh(x,y, tick_label = labels, color = colors)
plt.show()
```

```python
import numpy as np
x = [1,2]
y1, y2, y3 = [1,2],[2,5],[3,9]
fig = plt.figure()
ax = fig.add_subplot(111)
w = 0.2
ax.bar(x,y1,width=w, label = "english")
ax.bar(np.array(x)+w,y2,width=w, label = "math")
ax.bar(np.array(x)+w*2,y3,width=w, label = "korean")
ax.legend()
plt.show()
```
